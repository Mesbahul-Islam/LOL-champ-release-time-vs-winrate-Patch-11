```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing the data which I had cleaned using excel beforehand
df = pd.read_csv('LOL_champions_stats_edited.csv',encoding = 'latin1')
df
```

```python
#Selecting the necessary data
df2 = df[['Champion Name', 'Date_Release','Last_change']]
```

```python
#Renaming the columns and importing second set of data which i also cleaned in excel
df2.rename(columns = {'Champion Name': 'Name'})
df_new = pd.read_csv('League of Legends Champion Stats 12.3_edited.csv')
```

```python
#Setting the first row as the column labels as it was not before 
df_new.columns = df_new.iloc[0]
df_new.drop(axis = 0, labels = 0,inplace = True)
```

```python
#Fixing the column name 
df2.rename(columns={'Champion Name':'Name'},inplace=True)
```

```python
#Removing null values
df_new.dropna()
```

```python
#The original dataset df_new had win rates for same champions played in different lanes. So im dropping 
# the duplicates for the second occurance  
df_new.drop_duplicates(subset=['Name'],inplace=True)
```

```python
#Dropping the null values again
df_new.dropna()
```

```python
#Merging the datasets on name
df_final = df_new.merge(df2).copy()
```

```python
df_final
```

```python
df_final.head()
```

```python
#Dropping unnessary columns
df_final.drop(columns=['Class','Role','Tier','Score','Trend','Role %','KDA','Last_change'],inplace = True)
```

```python
df_final
```

```python
#The date of release column has strings. Converting the column to datetime format
df_final['Date_Release'] = pd.to_datetime(df_final['Date_Release'],dayfirst=True)
df_final
```

```python
#The percentage columns are written using strings. So im removing the periods and percentage so that I can convert
#the strings to float types
df_final['Ban %'] = (df_final['Ban %'].str.split()).apply(lambda x: (x[0].replace('.','')))
df_final['Ban %'] = (df_final['Ban %'].str.split()).apply(lambda x: (x[0].replace('%','')))
```

```python
#Converting to float type and dividing by 100 for percentages
df_final['Ban %'] = df_final['Ban %'].astype(float)/100
```

```python
#Creating the lineplot. In vscode(using it for experimental purposes. Good so far but its easier in jupyter notebook)
#the axis labels and axis ticks cannot be seen as they are black on a grey background. So i made it easier by making
#the colour red
plt.figure(figsize=(20,13))
fig1 = sns.lineplot(x='Date_Release',y='Win %',data = df_final)
fig1.xaxis.label.set_color('red')
fig1.spines['bottom'].set_color('red')
fig1.spines['top'].set_color('red')
fig1.tick_params(axis='both', colors='red')
fig1.yaxis.label.set_color('red')
```
